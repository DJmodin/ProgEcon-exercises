```python
a = 1
def f(x):
    return x+a # a is global variable
def g(x,a=1):
    return x+a # a is local variable
print(f(1))
print(g(1))
a = 2
print(f(1))
print(g(1))
```

```python
class SquareClass:
    def __init__(self ,length ,width):
        self.length = length
        self.width = width
    def size(self):
        return self.length*self.width
square = SquareClass (2,2)
print(square.size())
```
